```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument

sentences = [
    "NLP is interesting",
    "I love learning about AI",
    "Machine learning is powerful"
]

tagged = [TaggedDocument(words=s.split(), tags=[str(i)]) for i, s in enumerate(sentences)]

model = Doc2Vec(tagged, vector_size=50, min_count=1, epochs=40)

print(model.infer_vector(["NLP", "is", "fun"]))
```
